###### Существует 3 различные категории карт цветов:

- ``Последовательные`` - состоят из одной непрерывной последовательности цветов (``binary`` или ``viridis``...).
- ``Дивергентные`` - обычно содержат 2 хорошо различимых цвета, отражающих положительные и отрицательные отклонения от среднего значения(``BdBu`` или ``PuOr``).
- ``Качественные`` - в них цвета смешиваются без какого-либо четкого порядка (``rainbow`` или ``jet``).
- ``Циклические`` - изменение яркости двух разных цветов, которые встречаются в середине и начинаются/заканчиваются ненасыщенным цветом; следует использовать для значений, которые повторяются в конечных точках, таких как фазовый угол, направление ветра или время суток.

```python
from matplotlib.colors import LinearSegmentedColormap

def grayscale_cmap(cmap):
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    # преобразуем RGBA в воспринимаемую глазом светимость серого цвета
    # cp. alienryderflex.com/hsp.html
    RGB_weight = [0.299, 0.587, 0.114]
    luminace = np.sqrt(np.dot(colors[:, :3] ** 2, RGB_weight))
    colors[:, :3] = luminace[:, np.newaxis]
    return LinearSegmentedColormap.from_list(cmap.name + '_grey', colors, cmap.N)
```

```python
def view_colormap(cmap):
    """Рисует карту цветов в эквивалентных оттенках серго"""
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    cmap = grayscale_cmap(cmap)
    grayscale = cmap(np.arange(cmap.N))
    
    fig, ax = plt.subplots(2, figsize=(6, 2),
                           subplot_kw=dict(xticks=[], yticks=[]))
    ax[0].imshow([colors], extent=[0, 10, 0, 1])
    ax[1].imshow([grayscale], extent=[0, 10, 0, 1])
```

```python
view_colormap('jet')
# неравномерная яркость
# в книге сказано, что jet карта по умолчанию до версии 2.0
# а потом стал viridis
# но похоже теперь опять jet
```

## Ограничения и расширенные возможности по использованию цветов

Сами по себе шкалы цветов - просто экземпляры класса ``plt.Axes``, поэтому для них можно использовать трюки связанные с форматированием осей координат и делений на них.

```python
# создаем шум размером 1% от пикселей изображения
speckles = (np.random.random(I.shape) < 0.01)
I[speckles] = np.random.normal(0, 3, np.count_nonzero(speckles))

plt.figure(figsize=(10, 3.5))

plt.subplot(1, 2, 1)
plt.imshow(I, cmap='RdBu')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(I, cmap='RdBu')
plt.colorbar(extend='both') # рисуем стрелки
plt.clim(-1, 1) # устанавливаем диапазон
```

- 1) На левом рисунке зашумленные пиксели влияют на пределы диапазонов цветов
- 2) На правом рисунке мы задаем пределы диапазона цветов в ручную и добавляем стрелки, указывающие на значения выходящие за эти пределы.

```python
# создаем шум размером 1% от пикселей изображения
speckles = (np.random.random(I.shape) < 0.01)
I[speckles] = np.random.normal(0, 3, np.count_nonzero(speckles))

plt.figure(figsize=(10, 3.5))

plt.subplot(1, 2, 1)
plt.imshow(I, cmap='RdBu')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(I, cmap='RdBu')
plt.colorbar()
plt.clim(-1, 1)
```

### Пример: рукопсиные цифры

```python
# Загружаем изображения цифр от 0 до 5 и визуализируем некоторые
from sklearn.datasets import load_digits
digits = load_digits(n_class=6)

fig, ax = plt.subplots(8, 8, figsize=(6, 6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
projection = iso.fit_transform(digits.data)
```

```python
plt.scatter(projection[:, 0], projection[:, 1], lw=0.1,
            c=digits.target, cmap=plt.cm.get_cmap('cubehelix', 6))
plt.colorbar(ticks=range(6), label='Цифра')
plt.clim(-0.5, 5.5)
# многообразие 5 и 3 местами пересекается, что говорит о том, что отличить их друг от друга бывает не просто
```
