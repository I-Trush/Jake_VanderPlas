# Простые линейные графики

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure() # создание экземпляра рисунка (подложки/контейнера)
ax = plt.axes()  # создание экземпляра системы координат
```

В Matplotlib рисунок ``plt.figure()`` можно рассматривать как ``контейнер`` содержащий все объекты, представляющие:
- систему координат, ``ax = plt.axes`` - ограничивающий прямоугольник с делениями и метками, в котором будет находиться график
- графику (график) ``ax.plot()``
- текст
- метки

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

### Настройка графика: цвета и стили линий

```python
plt.plot?
```

```python
plt.plot(x, np.sin(x - 0), color='m')
# **Colors**

# The supported color abbreviations are the single letter codes

# =============    ===============================
# character        color
# =============    ===============================
# ``'b'``          blue
# ``'g'``          green
# ``'r'``          red
# ``'c'``          cyan
# ``'m'``          magenta
# ``'y'``          yellow
# ``'k'``          black
# ``'w'``          white
# =============    ===============================

# and the ``'CN'`` colors that index into the default property cycle.

# If the color is the only part of the format string, you can
# additionally use any  `matplotlib.colors` spec, e.g. full names
# (``'green'``) or hex strings (``'#008000'``).
```

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, np.sin(x - 0), color='m', linestyle='--')
# **Line Styles**

# =============    ===============================
# character        description
# =============    ===============================
# ``'-'``          solid line style
# ``'--'``         dashed line style
# ``'-.'``         dash-dot line style
# ``':'``          dotted line style
# =============    ===============================

# Example format strings::

#     'b'    # blue markers with default shape
#     'or'   # red circles
#     '-g'   # green solid line
#     '--'   # dashed line with default color
#     '^k:'  # black triangle_up markers connected by a dotted line
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# сокращения
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

## Настройка графика, пределы осей координат
``plt.xlim()`` and ``plt.ylim()``:

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.xlim(-5, 11)
plt.ylim(-1.5, 1.5)
```

```python
# для зеркального отображения осей
plt.plot(x, np.sin(x), color='blue')
plt.xlim(11, -5)
plt.ylim(1.5, -1.5)
```

Удобный метод ``plt.axis()`` (не путать с plt.axes)
- позволяет задавать пределы осей списком ``[xmin, xmax, ymin, ymax]``

```python
plt.plot(x, np.sin(x), color='m')
plt.axis([-2, 11, -2, 1.5])
```

``plt.axis()`` позволяет подогнать границы в упор

```python
plt.plot(x, np.sin(x), color='m')
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x), color='m')
plt.axis('equal')
```

```python
# plt.axis?
#     Value    Description
#     ======== ==========================================================
#     'on'     Turn on axis lines and labels. Same as ``True``.
#     'off'    Turn off axis lines and labels. Same as ``False``.
#     'equal'  Set equal scaling (i.e., make circles circular) by
#              changing axis limits. This is the same as
#              ``ax.set_aspect('equal', adjustable='datalim')``.
#              Explicit data limits may not be respected in this case.
#     'scaled' Set equal scaling (i.e., make circles circular) by
#              changing dimensions of the plot box. This is the same as
#              ``ax.set_aspect('equal', adjustable='box', anchor='C')``.
#              Additionally, further autoscaling will be disabled.
#     'tight'  Set limits just large enough to show all data, then
#              disable further autoscaling.
#     'auto'   Automatic scaling (fill plot box with data).
#     'image'  'scaled' with axis limits equal to data limits.
#     'square' Square plot; similar to 'scaled', but initially forcing
#              ``xmax-xmin == ymax-ymin``.
```

## Метки на графиках

```python
plt.plot(x, np.sin(x), color='m')
plt.title('синус') # название графика

plt.xlabel('x')
plt.ylabel('sin(x)')
```

Для удобного создания легенды для графика ``plt.legend()``

```python
plt.plot(x, np.sin(x), color='m', label='sin(x)')
plt.plot(x, np.cos(x), color='m', label='cos(x)')
```

```python
plt.plot(x, np.sin(x), color='m', label='sin(x)')
plt.plot(x, np.cos(x), color='m', label='cos(x)')
plt.axis('equal')
plt.legend()
```

## Нюансы
Совпадающие методы:
- ``plt.plot()`` → ``ax.plot()``
- ``plt.legend()`` → ``ax.legend()``
- и т.д.


Не совпадающие методы:
- ``plt.xlabel()``  → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()``  → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``

###### В объектно ориентированном интерфейсе удобно воспользоваться ф-ией ``ax.set()``

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(
    xlim=(0, 10), 
    ylim=(-2, 2),
    xlabel='x', ylabel='sin(x)',
    title='A Simple Plot'
)
```
