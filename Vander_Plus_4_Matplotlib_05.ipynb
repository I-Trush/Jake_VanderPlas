# Графики плотности и контурные графики

Иногда удобно отображать 3d графики на плоскости с помощью контуров и областей, кодированныз разными цветами. Для этого существуют 3 ф-ии библиотеки ``Matplotlib``:
- ``plt.contour`` - для контурных графиков
- ``plt.contourf`` - для контурных графиков с заполнением
- ``plt.imshow`` - для отображения изображений

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

###### Визуализация трехмерной ф-ии
z = f(x,y)

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y*x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
x
```

```python
X, Y = np.meshgrid(x, y)
X
```

```python
Z = f(X, Y)
```

```python
Z
```

```python
plt.contour(X, Y, Z, colors='black')
# отрицательные значения - штрихи
```

Можно кодировать линии различными цветами, задав карты цветом ``cmap``

```python
plt.contour(X, Y, Z, 20, cmap='inferno')
plt.colorbar()
# 20 - указали, что хотим больше линий, поделив данные на 20 промежутков
```

```python
plt.contour(X, Y, Z, 15, cmap='RdGy')
plt.colorbar()
```

###### Лучшее решение - ``plt.show()``  - ф-ия интерпретирующая двумерную сетка, как изображение

```python
plt.imshow(Z, extent=[0, 5 , 0, 5], origin='lower', cmap='coolwarm')
plt.colorbar()
# как-то больно пиксельно
```

###### Иногда бывает удобно комбинировать контурный график, с  графиком изображением

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5 , 0, 5], origin='lower', cmap='coolwarm', interpolation='gaussian')
plt.colorbar()
```

```python
plt.figure(figsize=(16,10))  # порядок имеет значение
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5 , 0, 5], origin='lower', cmap='coolwarm', interpolation='gaussian')
plt.colorbar()
```
