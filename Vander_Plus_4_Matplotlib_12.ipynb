# Пользовательские настройки Matplotlib: конфигурации и таблицы стилей

## Выполнение пользовательских настроек графиков вручную

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
```

```python
x = np.random.randn(1000)
plt.hist(x)
```

```python
ax = plt.axes()
ax.set_facecolor('#E6E6E6') # серый фон
ax.set_axisbelow(True) # сетка за гистограммой

# Рисуем сплошные белые линии сетки
plt.grid(color='w', linestyle='solid')

# Скрываем основные линии осей координат
for spine in ax.spines.values():
    spine.set_visible(False)
    
# Скрываем деления сверху и справа
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# Осветляем цвет делений и меток
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
    
# Задаем цвет заливки и границ гистограммы
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666')
```

## Изменяем значения по умолчанию: ``rcParams``

Конфигурацию можно настроить при помощи ``plt.rc``

```python
# сперва скопируем словарь с default настройками, чтоб было удобно откатиться
IPython_default = plt.rcParams.copy()
# IPython_default = plt.rcParamsDefault существует такой метод
```

```python
from matplotlib import cycler

colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])

plt.rc('axes', 
       facecolor='#E6E6E6', 
       edgecolor='none',
       axisbelow=True, 
       grid=True, 
       prop_cycle=colors)

plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)
```

## Таблицы стилей

Стили находятся в файле ``mplstyle``

Имеющиеся стили перечислены в ``plt.style.available``

```python
plt.style.available
```

```python
# чтоб установить стиль
plt.style.use('seaborn-dark-palette')
```

```python
for i in range(4):
    plt.plot(np.random.rand(10))
```

###### Можно установить стиль временно:

```python
with plt.style.context('fivethirtyeight'):
    for i in range(4):
        plt.plot(np.random.rand(10))
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```
